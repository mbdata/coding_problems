It is called ordinary because historically it was the first statistical procedure to be developed, around the time of the 1800s. **52) Why do vicious feedback loops matter in data analytics?**

Vicious feedback loops are important in data analytics because they allow us to improve our model while we apply it in real life analysis. These loops can adversely affect our analytical models when the outputs fed back into the model should not have been inputted. We can to guard against these vicious loops by retraining our model once in a while.

**53) What are the 10 simple rules Zook et. al. Suggest to follow when doing big data analytics?**

The 10 Rules are:
1. Acknowledge that data are people and can do harm
2. Recognize that privacy is more than a binary value
3. Guard against the reidentification of your data
4. Practice ethical data sharing
5. Consider the strengths and limitations of your data; big does not automatically mean better
6. Debate the tough, ethical choices
7. Develop a code of conduct for your organization, research community, or industry
8. Design your data and systems for auditability
9. Engage with the broader consequences of data and analysis practices
10. Know when to break these rules
